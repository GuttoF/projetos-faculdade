### Imports e Funções

```python
import numpy as np
import pandas as pd
```

```python
def calculate_layer_output(inputs, weights, biases):
    layer_output = []
    for input_set in inputs:
        neuron_outputs = []
        for neuron_weights, bias in zip(weights, biases):
            neuron_output = sum(input_value * weight for input_value, weight in zip(input_set, neuron_weights)) + bias
            neuron_outputs.append(neuron_output)
        layer_output.append(neuron_outputs)
    return layer_output
```

```python
inputs = np.array([[1, 2, 3, 2.5],
                   [2, 5, -1, 2],
                   [-1.5, 2.7, 3.3, -0.8]
                  ])

weights = np.array([[0.2, 0.8, -0.5, 1],
                    [0.5, -0.91, 0.26, -0.5],
                    [-0.26, -0.27, 0.17, 0.87]
                  ])

biases = np.array([2, 3, 0.5])

weights2 = np.array([[0.1, -0.14, 0.5],
                     [-0.5, 0.12, -0.33],
                     [-0.44, 0.73, -0.13]
                    ])

biases2 = np.array([-1, 2, -0.5])


inputs_manual = [[1, 2, 3, 2.5],
                 [2, 5, -1, 2],
                 [-1.5, 2.7, 3.3, -0.8]]

weights_manual = [[0.2, 0.8, -0.5, 1],
                  [0.5, -0.91, 0.26, -0.5],
                  [-0.26, -0.27, 0.17, 0.87]]

biases_manual = [2, 3, 0.5]

weights2_manual = [[0.1, -0.14, 0.5],
                   [-0.5, 0.12, -0.33],
                   [-0.44, 0.73, -0.13]]

biases2_manual = [-1, 2, -0.5]
```

### Resolvendo por multiplicação de matrizes

```python
layer1_output = np.dot(inputs, weights.T) + biases

output = np.dot(layer1_output, weights2.T) + biases2

df = pd.DataFrame(output, columns=["Neurônio 1", "Neurônio 2", "Neurônio 3"])
df.index.name = "Output"
df.style.format("{:.3f}")
df
```

### Resolvendo manualmente

```python
layer1_output_manual = calculate_layer_output(inputs_manual, weights_manual, biases_manual)
output_manual = calculate_layer_output(layer1_output_manual, weights2_manual, biases2_manual)

df_manual = pd.DataFrame(output_manual, columns=["Neurônio 1", "Neurônio 2", "Neurônio 3"])
df_manual.index.name = "Output"
df_manual.style.format("{:.3f}")
df_manual
```

### Explicação

O output mostra as respostas da rede neural para cada conjunto de dados que ela recebeu. Cada linha representa uma  amostra diferente, e cada coluna mostra a resposta de um neurônio específico na camada final. Esses números representam a “opinião” da rede sobre cada amostra, com base no que ela aprendeu.
